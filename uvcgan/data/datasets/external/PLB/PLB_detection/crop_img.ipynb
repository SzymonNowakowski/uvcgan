```python
import os
import cv2
from tqdm import tqdm
```

```python
def crop_yolo_bboxes(images_dir, labels_dir, output_dir):

    os.makedirs(output_dir, exist_ok=True)

    for label_file in os.listdir(labels_dir):
        if not label_file.endswith(".txt"):
            continue

        label_path = os.path.join(labels_dir, label_file)
        image_name = os.path.splitext(label_file)[0]

        for ext in [".jpg", ".png", ".jpeg"]:
            image_path = os.path.join(images_dir, image_name + ext)
            if os.path.exists(image_path):
                break
        else:
            print(f"No image found for {label_file}")
            continue

        img = cv2.imread(image_path)
        if img is None:
            print(f"Failed to load image: {image_path}")
            continue

        h, w, _ = img.shape

        with open(label_path, "r") as f:
            lines = f.readlines()

        for i, line in enumerate(lines):
            parts = line.strip().split()
            if len(parts) < 5:
                continue

            cls, x_center, y_center, bw, bh = map(float, parts[:5])

            x_center *= w
            y_center *= h
            bw *= w
            bh *= h

            x1 = int(x_center - bw / 2)
            y1 = int(y_center - bh / 2)
            x2 = int(x_center + bw / 2)
            y2 = int(y_center + bh / 2)

            x1, y1 = max(0, x1), max(0, y1)
            x2, y2 = min(w, x2), min(h, y2)

            crop = img[y1:y2, x1:x2]

            output_path = os.path.join(output_dir, f"{image_name}_obj{i+1}.jpg")
            cv2.imwrite(output_path, crop)
            print(f"Zapisano: {output_path}")

    print("Finished processing all files.")
```

```python
crop_yolo_bboxes(
    images_dir="/content/drive/MyDrive/Krysztaly_plb/02_training/train_1_plb/pred_unique_jpgs_1601_files_907cd50/predict2",
    labels_dir="/content/drive/MyDrive/Krysztaly_plb/02_training/train_1_plb/pred_unique_jpgs_1601_files_907cd50/predict2/labels",
    output_dir="/content/drive/MyDrive/Krysztaly_plb/02_training/train_1_plb/pred_unique_jpgs_1601_files_907cd50/crop"
)
```
